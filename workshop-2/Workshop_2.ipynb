```python
class Engine:
  def __init__(self, name_engine: str, potency_engine: int, weight: float, stability, power, dimensions, torque, maximum_speed, engine_cost):
    self.name = name_engine
    self.potency = potency_engine
    self.stability = stability
    self.power = power
    self.weight = weight
    self.dimensions = dimensions
    self.torque = torque
    self.maximum_speed = maximum_speed
    self.engine_cost = engine_cost

class GasEngine(Engine):
  def __init__(self, stability, power, weight, dimensions, torque, maximum_speed, engine_cost):
    super().__init__( stability, power, weight, dimensions, torque, maximum_speed, engine_cost)
    self.combustible_type = "Gas"

class ElectricEngine(Engine):
  def __init__(self, stability, power, weight, dimensions, torque, maximum_speed, engine_cost):
    super().__init__( stability, power, weight, dimensions, torque, maximum_speed, engine_cost)
    self.combustible_type = "Electric"

class EngineFactory:
  def createGasEngine(name_engine: str, potency_engine: int, weight: float, stability, power, dimensions, torque, maximum_speed) -> Engine:
    pass

  def createElectricEngine(name_engine: str, potency_engine: int, weight: float, stability, power, dimensions, torque, maximum_speed) -> Engine:
    pass

class LowEngineFactory(EngineFactory):
  def createGasEngine(name_engine: str, potency_engine: int, weight: float, stability, power, dimensions, torque, maximum_speed) -> Engine:
    return GasEngine(stability, power, weight, dimensions, torque, maximum_speed,
                     "Low")

  def createElectricEngine(name_engine: str, potency_engine: int, weight: float, stability, power, dimensions, torque, maximum_speed) -> Engine:
    return ElectricEngine(stability, power, weight, dimensions, torque, maximum_speed, "Low")

class HighEngineFactory(EngineFactory):
  def createGasEngine(name_engine: str, potency_engine: int, weight: float, stability, power, dimensions, torque, maximum_speed) -> Engine:
    return GasEngine(stability, power, weight, dimensions, torque, maximum_speed, "High")

  def createElectricEngine(name_engine, potency_engine, weight, stability, power, dimensions, torque, maximum_speed) -> Engine:
    return ElectricEngine(stability, power, weight, dimensions, torque, maximum_speed, "High")
```

```python
class Vehicle:
  """
    This class represent the structure and behaviors of any vehicle
  """

  # Constructor
  def __init__(self, engine: Engine, trade: str, model: str, year: int):

    # Attributes of the class
    self.engine = engine
    self.model = model
    self.trade = trade
    self.year = year
    self.gas_consumption = self.cal_gas()



class Car(Vehicle):
  """
    This class represent the structure and behaviors of any Car
  """
  def __init__(self, engine, transmision, chasis, model, year):
    super().__init__(engine, model, year)
    self.chasis = chasis
    self.transmision = transmision
    self.combustible_type = self.engine.engine_type


class Truck(Vehicle):
  """
    This class represent the structure and behaviors of any Truck
  """
  def __init__(self, engine, chasis, model, year):
    super().__init__(engine, chasis, model, year)


  def cal_gas(self) -> float:

    """
      This method calculated and returns the gas consumption

      Returns:
      . (str) gas consumption
    """
    if(self.chasis == "A"):
      val_chasis = 0.3
    else:
      val_chasis = 0.5

    return (1.1*self.engine.potency)+(0.2*self.engine.weight)-val_chasis

class Yacht(Vehicle):
  """
    This class represent the structure and behaviors of any Yacht
  """
  def __init__(self, engine, length, weight, chasis, model, year):
    super().__init__(engine, chasis, model, year)
    self.length = length
    self.weight = weight

class Motorcycle(Vehicle):
  """
    This class represent the structure and behaviors of any Motorcycle
  """
  def __init__(self, engine, chasis, model, year):
    super().__init__(engine, chasis, model, year)
    self.chasis = chasis

class CatalogVehicles:
  def create_vehicle(self, type_) -> Vehicle:
    if type_ == "Car":
      return "Car"
    elif type_ == "Yacht":
      return "Yacht"
    elif type_ == "Truck":
      return "Truck"
    elif type_ == "Motorcycle":
      return "Motorcycle"
```
